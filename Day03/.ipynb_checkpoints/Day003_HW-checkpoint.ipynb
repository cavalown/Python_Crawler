### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict

res = "http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"


with open('./data/64_72hr_CH.xml','r') as f:
    file_dict = dict(xmltodict.parse(f.read()))

# file_dict = dic(xmltodict.parse())
# print(file_dict)
kaohsiung = file_dict['cwbopendata']['dataset']['locations']['location']
print(len(kaohsiung))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for district in kaohsiung:
#     print(district)
    location = district['locationName']
    record_time = district['weatherElement'][0]['time'][0]['dataTime']
    temperature = district['weatherElement'][0]['time'][0]['elementValue']['value']
    print(f'地區：{location}，時間：{record_time}，溫度：{temperature}度(攝氏)')
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

place = kaohsiung[0]['locationName']
print(f'高雄市的『{place}』')

for item in kaohsiung[0]['weatherElement'][0]['time']:
    each_time = item['dataTime']
    each_temp = item['elementValue']['value']
    print(f'時間：{each_time}，溫度：{each_temp}度（攝氏）')
```
